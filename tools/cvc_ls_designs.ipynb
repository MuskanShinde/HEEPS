### Lyot Stop designs with 8% inner edge oversize
##### ls_dRext=0.0291, ls_dRint=0.08, ls_dRspi=0.0317

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import os.path
import heeps
import heeps.util.img_processing as impro
from heeps.util.coord import cart_coord, polar_coord
import astropy.units as u
```

### Circular Lyot Stop

```python
psf_ON = fits.getdata('ls_LM_CVC_0.08_285_onaxis_PSF_L_CVC.fits')
plt.imshow(np.log10(psf_ON), origin='lower')
plt.title('on-axis PSF')
plt.colorbar()
plt.clim(-4, -14)
```

```python
psf_OFF = fits.getdata('ls_LM_CVC_0.08_285_offaxis_PSF_L_CVC.fits')
plt.imshow(np.log10(psf_OFF), origin='lower')
plt.title('off-axis PSF')
plt.colorbar()
plt.clim(-2, -16)
```

```python
# radial profile
c=psf_OFF.shape[0]//2
rad_prof = impro.get_radial_profile(psf_OFF, (c,c), nbin=1)
pscale = 5.47 #mas
lam = 3.8112E-06 #m
diam_ext = 36.905 #m
radius = (np.arange(c+1))*pscale*u.mas.to('rad')/(lam/diam_ext)
plt.grid(True)
plt.plot(radius, rad_prof, )
plt.title('Radial Profile off-axis PSF')
plt.ylabel('Intensity')  
plt.xlabel('Radius $[\lambda/D]$')
plt.xlim(0.0, 5.0)
plt.ylim(0, 0.025)

#flux in aperture of 1 lambda/D
r, t = polar_coord(psf_OFF.shape[0])
r_1lamD = 1/radius[-1]
mask_1lamD = (r < r_1lamD)
flux_1lamD = np.sum(mask_1lamD*psf_OFF)/np.sum(psf_OFF)
print('Flux in aperture of 1 lambda/D = %s'%flux_1lamD)
```
